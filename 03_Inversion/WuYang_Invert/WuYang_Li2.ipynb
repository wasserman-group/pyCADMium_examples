```python
import numpy as np
import matplotlib.pyplot as plt

from CADMium import Pssolver
from CADMium import Psgrid
from CADMium import Partition
from CADMium import Inverter

a = 5.122/2
#Nuclear charge for fragments A and B
Za, Zb = 3,3
#Set polarization 1-Unpolarized, 2-Polarized
pol = 2
#Fragment a electrons [alpha, beta]
Nmo_a = [[2,1]] #Number of molecular orbitals to calculate
N_a   = [[2,1]]
#Ensemble mix
nu_a = 1
#Fragment b electrons
Nmo_b = [[2,1]]
N_b   = [[2,1]]
#Ensemble mix
nu_b = 1

#Molecular elctron configuration
Nmo_m = [[3,3]]
N_m   = [[3,3]]
 
opt = {   
          "kinetic_part_type" : "inversion",
          "AB_SYM" : True,
          "ENS_SPIN_SYM" : True,
          #Inversion flags:
          "invert_type" : "wuyang",
          #Scf Flag
          "Verbose" : True,
            }


#Set up grid
NP = 7
NM = [4,4]
L = np.arccosh(15/a)
loc = np.array(range(-4,5)) #Stencil outline

grid = Psgrid(NP, NM, a, L, loc)
grid.initialize()

part = Partition(grid, Za, Zb, pol, Nmo_a, N_a, nu_a, Nmo_b, N_b, nu_b, opt)

#Setup inverter object
mol_solver = Pssolver(grid, Nmo_m, N_m)
part.inverter = Inverter(grid, mol_solver, opt)

part.scf()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from CADMium import Pssolver
from CADMium import Psgrid
from CADMium import Partition
from CADMium import Inverter

a = 5.122/2
#Nuclear charge for fragments A and B
Za, Zb = 3,3
#Set polarization 1-Unpolarized, 2-Polarized
pol = 1
#Fragment a electrons [alpha, beta]
Nmo_a = [[3]] #Number of molecular orbitals to calculate
N_a   = [[3]]
#Ensemble mix
nu_a = 1
#Fragment b electrons
Nmo_b = [[3]]
N_b   = [[3]]
#Ensemble mix
nu_b = 1

#Molecular elctron configuration
Nmo_m = [[6]]
N_m   = [[6]]
 
opt = {   
          "kinetic_part_type" : "inversion",
          "AB_SYM" : True,
          "ENS_SPIN_SYM" : False,
          #Inversion flags:
          "invert_type" : "wuyang",
          #Scf Flag
          "Verbose" : True,
            }


#Set up grid
NP = 6
NM = [6,6]
L = np.arccosh(15/a)
loc = np.array(range(-4,5)) #Stencil outline

grid = Psgrid(NP, NM, a, L, loc)
grid.initialize()

part = Partition(grid, Za, Zb, pol, Nmo_a, N_a, nu_a, Nmo_b, N_b, nu_b, opt)

#Setup inverter object
mol_solver = Pssolver(grid, Nmo_m, N_m)
part.inverter = Inverter(grid, mol_solver, opt)

part.scf()
```
