```python
import numpy as np
import matplotlib.pyplot as plt

from CADMium import Pssolver
from CADMium import Psgrid
from CADMium import Partition
from CADMium import Inverter

a = 6/2
#Nuclear charge for fragments A and B
Za, Zb = 2,2
#Set polarization 1-Unpolarized, 2-Polarized
pol = 1
#Fragment a electrons [alpha, beta]
Nmo_a = [[1]] #Number of molecular orbitals to calculate
N_a   = [[2]]
#Ensemble mix
nu_a = 1
#Fragment b electrons
Nmo_b = [[1]]
N_b   = [[2]]
#Ensemble mix
nu_b = 1

#Molecular elctron configuration
Nmo_m = [[2]]
N_m   = [[4]]

opt = {   
          "kinetic_part_type" : "inversion",
          "AB_SYM" : True,
          "ENS_SPIN_SYM" : False,
            }


#Set up grid
NP = 7
NM = [4,4]
L = np.arccosh(15/a)
loc = np.array(range(-4,5)) #Stencil outline

grid = Psgrid(NP, NM, a, L, loc)
grid.initialize()

part = Partition(grid, Za, Zb, pol, Nmo_a, N_a, nu_a, Nmo_b, N_b, nu_b, opt)

#Setup inverter object
mol_solver = Pssolver(grid, Nmo_m, N_m)
part.inverter = Inverter(grid, mol_solver, {"disp"         : True,
                                            "invert_type"  : "wuyang", 
                                            "ab_sym"       : True,
                                            "ens_spin_sym" : False})

part.scf({"disp" : True})
```

```python
part.inverter.solver
```

```python

full, x,y = grid.plotter(part.V.vp[:,0])
x_axis,vp = grid.axis_plot(part.V.vp[:,0])
```

```python
fig = plt.figure(dpi=150)
plt.contourf(x,y,full, levels=30, cmap="viridis")
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.colorbar()
```

```python
fig = plt.figure(dpi=120)
plt.plot(x_axis,vp)
plt.xlim(0,5)
```

```python
x, na_grid = grid.axis_plot(part.na_frac)
_, nb_grid = grid.axis_plot(part.nb_frac)
_, nf_grid = grid.axis_plot(part.nf)
```

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, na_grid, lw=5)
ax.plot(x, nb_grid, lw=5)
ax.plot(x, nf_grid, ls=":", lw=5)
```

```python
#Let us look at the ungerade orbital first:

norm_1 = 1/ grid.integrate( (part.na_frac**(0.5) - part.nb_frac**(0.5))**2 )**(0.5)
norm_2 = 1/ grid.integrate( (part.na_frac**(0.5) + part.nb_frac**(0.5))**2 )**(0.5)


ungerad = norm_1 * (part.na_frac**(0.5) - part.nb_frac**(0.5))
x, ungerad_gd = grid.axis_plot(ungerad)

gerad = norm_2 * (part.nf/2 - ungerad**2)**(0.5)
x, gerad_gd = grid.axis_plot(gerad)
```

```python
fig, ax = plt.subplots()
ax.plot(x, ungerad_gd, label="Ungerad?")
ax.plot(x, gerad_gd, label="Gerad?")

ax.legend()
```
